Neste notebook, realizamos as seguintes etapas:

1. **Configura칞칚o do Ambiente**  
   Selecionamos o cat치logo `100cep_gateway` e o schema `staging` para garantir que todos os dados e objetos criados estejam organizados no ambiente correto.

2. **Cria칞칚o de Volume no Unity Catalog**  
   Criamos um volume chamado `imdb` para armazenar arquivos de dados de forma segura e centralizada no Unity Catalog.

3. **Instala칞칚o e Importa칞칚o de Bibliotecas**  
   Instalamos a biblioteca `kagglehub` para facilitar o download de datasets diretamente do Kaggle.

4. **Download e C칩pia do Dataset**  
   Utilizamos o `kagglehub` para baixar o dataset "olistbr/brazilian-ecommerce" do Kaggle. Em seguida, copiamos os arquivos baixados para o volume criado no Unity Catalog, garantindo que os dados estejam dispon칤veis para an치lises futuras.

Essas etapas preparam o ambiente para an치lises avan칞adas utilizando dados reais de e-commerce brasileiro, com governan칞a e seguran칞a providas pelo Unity Catalog.

```python
import shutil
import os

# Download latest version
path = kagglehub.dataset_download(
    "olistbr/brazilian-ecommerce"
)

uc_volume_path = "/Volumes/100cep_gateway/staging/imdb"

os.makedirs(uc_volume_path, exist_ok=True)

for file_name in os.listdir(path):
    src_file = os.path.join(path, file_name)
    dst_file = os.path.join(uc_volume_path, file_name)
    # Copy and overwrite if exists
    shutil.copy2(src_file, dst_file)

print(
    f"Arquivo copiado para Unity Catalog volume em: {uc_volume_path}"
)
```
